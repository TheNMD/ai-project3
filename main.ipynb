# Important values in a radar image [1], [2]

#### 1. Reflectivity (Echo Intensity) (Measured in dBZ) - Precipitation
##### 1.1. Base Reflectivity & Color Coding
- Reflectivity values represent the intensity of precipitation.
- Higher dBZ values generally indicate more intense rainfall.
- Based on specific dBZ ranges, rainfall is usually categorized as:
    - Light rain: 10-20 dBZ (Colored light green). 
    - Moderate rain: 20-40 dBZ (Colored yellow).  
    - Heavy rain: above 40 dBZ (Colored red).

##### 1.2. Composite Reflectivity
- Comprises of maximum reflectivities measured from four radar tilt angles (0.5°, 1.45°, 2.40°, 
  and 3.35°)
- Composite reflectivity can reveal important storm structure features and intensity trends of storms
  when compared with base reflectivity.

##### 1.3. Reflectivity Patterns:
- Different reflectivity patterns can indicate various types of precipitation. 
- Convective cells with high reflectivity values may indicate thunderstorms and heavy rainfall.

#### 2. Doppler Radar (Measured in m/s or knots) - Wind
##### 2.1. Doppler Velocity & Color Coding
- The velocity of precipitation particles moving toward or away from the radar site.
- The magnitude of the Doppler velocity can be used to estimate the speed of the winds 
  associated with precipitation. 
- The sign indicates the direction of the velocity:
    - Positive values indicate motion toward the radar (radial inbound) (Colored red).
    - Negative values indicate motion away from the radar (radial outbound) (Colored blue).
- 1 knots = 1 nautical mile /h = 1.852 km/h.

##### 2.2. Velocity Patterns
- Doppler velocity patterns can be used to identify wind patterns.
- Some specific velocity pattern like rotation usually appear in severe weather phenomena like 
  thunderstorms.

#### 3. Radar Range Ring (nautical mile)
- Indicate the distance from the radar site.
- 1 nautical mile = 1.852 km.

#### 4. Timestamp (YYYY-MM-DD HH:MM:SS)
Indicate when the data was collected.

# Some labels for weather radar images:
1. Clear sky (Reflectivity; Reflectivity Patterns)
2. Rain (Reflectivity; Reflectivity Patterns; Doppler Velocity)
3. Heavy rain (Reflectivity Patterns; Reflectivity Patterns; Doppler Velocity)
4. Thunderstorm (Reflectivity Patterns; Reflectivity Patterns; Doppler Velocity)
5. Typhoon (Storm eye structure; Eyewall replacement cycle; Reflectivity; Reflectivity Patterns; Doppler Velocity)
